Data Preperation

```python
split0, split1, split2 = tfds.even_splits('train', n=3)
```

```python
ds_train, ds_info = tfds.load(
    'malaria',
    split=split0,
    shuffle_files=False,
    as_supervised=True,
    with_info=True,
    )
```

```python
ds_test = tfds.load(
    'malaria',
    split=split1,
    shuffle_files=False,
    as_supervised=True,
    with_info=True,
    )
```

```python
ds_val = tfds.load(
    'malaria',
    split=split2,
    shuffle_files=False,
    as_supervised=True,
    with_info=True,
    )
```

Show examples

```python
fig = tfds.show_examples(ds_train, ds_info, rows=4, cols=4)
```
